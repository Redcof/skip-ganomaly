```python
class Option:
    def __init__(self):
        self.dataset='sixray_sd3'
        self.dataroot='dataset/sample'       
        self.patht=''
        self.batchsize=1
        self.workers=8
        self.droplast=False
        self.isize=128
        self.nc=3
        self.nz=100
        self.ngf=64
        self.ndf=64
        self.extralayers=0
        self.device='gpu'
        self.gpu_ids=[0]
        self.ngpu=1
        self.name='experiment_name' 
        self.model = 'skipganomaly'
        self.display_server="http://localhost"
        self.display_port=8097
        self.display_id=0,
        self.display=False
        self.verbose= False
        self.outf='./output'
        self.manualseed=47
        self.abnormal_class='automobile'
        self.metric='roc'

        ##
        # Train
        self.print_freq=100
        self.save_image_freq=100
        self.save_test_images=False
        self.load_weights=False
        self.resume = 'output/skipganomaly/sixray_sd3/train/weights/last'
        self.phase='train'
        self.iter=0
        self.niter=15
        self.niter_decay=100
        self.beta1=0.5
        self.lr=0.0002
        self.w_adv=1
        self.w_con=50
        self.w_lat=1
        self.lr_policy='lambda'
        self.lr_decay_iters=50

        self.isTrain = False

opt = Option()
```

```python
data = load_data(opt)
len(data.train)
```

```python
model = load_model(opt, None)
```

```python
for idx, (X,y) in enumerate(data.train):
    fakeX = model.netg(X)
    img = (fakeX.detach().cpu().reshape((128,128,3)).numpy()*255).astype("uint8")
    plt.subplot(2, 5, idx+1)
    plt.imshow(img)
    
plt.show()
```
